```python
def size_to_kb(size):
    if size == 'Varies with device':
        return np.nan
    if 'M' in size:
        return float(size.replace('M', '')) * 1024
    if 'G' in size:
        return float(size.replace('G', '')) * 1024 * 1024
    if 'k' in size:
        return float(size.replace('k', ''))
    return float(size)

df['Size_KB'] = df['Size'].apply(size_to_kb)

# Convert Installs to numerical
df['Installs'] = df['Installs'].str.replace('[+,]', '', regex=True).astype(int)

# Convert Price to numerical
df['Price'] = df['Price'].str.replace('$', '').astype(float)
```

# Performing EDA 

## 1. Top 10 Apps with Maximum Installs

```python
df['Installs'] = df['Installs'].astype(str).str.replace('[+,]', '', regex=True).astype(int)  
top_10_apps = df.nlargest(10, 'Installs')[['App', 'Installs']]
print(top_10_apps)
top_10_apps = pd.DataFrame(top_10_apps)
```

## 2. Distribution of App Ratings

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'].dropna(), bins=20, kde=True)
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

## 3. Average Rating by Category

```python
plt.figure(figsize=(14, 8))
sns.barplot(x='Rating', y='Category', data=df, estimator=np.mean,ci=0)
plt.title('Average Rating by Category')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.show()
```

## 4. Free vs Paid Apps

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Type', data=df, palette = 'winter')
plt.title('Free vs Paid Apps')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

## 5. Content Rating Distrbution

```python
plt.figure(figsize=(10, 6))
df['Content Rating'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Content Rating Distribution')
plt.ylabel('')
plt.show()
```

## 6. Top 10 Apps with Most Reviews

```python
df['Reviews'] = df['Reviews'].astype(str).str.replace('[+,]', '', regex=True).astype(int)

# Group by 'App' and sum the reviews
grouped_df = df.groupby('App', as_index=False)['Reviews'].sum()

# Get the top 10 apps with the highest number of reviews
top_10_reviews = grouped_df.nlargest(10, 'Reviews')

print(top_10_reviews)
```

```python
top_10_reviews = grouped_df.nlargest(10, 'Reviews')

# Plot the results
plt.figure(figsize=(10, 6))
sns.barplot(x='Reviews', y='App', data=top_10_reviews)
plt.title('Top 10 Apps with Most Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('App')
plt.show()
```

## 7. Price Distribution of Paid Apps

```python
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Price'] > 0]['Price'], bins=20, kde=True)
plt.title('Price Distribution of Paid Apps')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

## 8. Average Size of Apps by Category

```python
plt.figure(figsize=(14, 8))
sns.barplot(x='Size_KB', y='Category', data=df, estimator=np.mean,ci=0)
plt.title('Average Size of Apps by Category')
plt.xlabel('Average Size (KB)')
plt.ylabel('Category')
plt.show()
```

## 9. Distribution of App Updates Over the Years

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'])
df.head()
```

```python
df['Year'] = df['Last Updated'].dt.year

# Count the number of updates per year
yearly_counts = df['Year'].value_counts().sort_index()

# Plot the data
plt.figure(figsize=(14, 8))
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values, marker='o')
plt.title('Distribution of App Updates Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Apps Updated')
plt.grid(True)
plt.show()
```

## 10. Top 10 Categories with Most Installs

```python
category_installs = df.groupby('Category')['Installs'].sum().nlargest(10).reset_index()
plt.figure(figsize=(14, 8))
sns.barplot(x='Installs', y='Category', data=category_installs)
plt.title('Top 10 Categories with Most Installs')
plt.xlabel('Total Installs')
plt.ylabel('Category')
plt.show()
```

##  14. Most Common Genres

```python
# Count the number of apps in each genre
genre_counts = df['Genres'].value_counts()

# Get the top 10 genres by count
top_10_genres = genre_counts.head(10)
top_10_genres = pd.DataFrame(top_10_genres)
top_10_genres =  top_10_genres.reset_index()
top_10_genres.rename(columns={'index': 'Genres','Genres':'Count_of _Genres'}, inplace=True)

# Display the top 10 genres
top_10_genres
```
